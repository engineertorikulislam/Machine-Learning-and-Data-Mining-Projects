```python
from sklearn import datasets,cluster
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x=datasets.load_iris().data
y=datasets.load_iris().target
y
```

```python
db=cluster.DBSCAN(eps=0.4,min_samples=4)#eps range for search area
```

```python
db.fit(x)
```

```python
pred=db.labels_ #-1 for a sample ,means it is noise
pred
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy = {round(accuracy_score(y,pred)*100,2)}%")
```

```python
figsize=(10,10)
plt.figure(figsize=(10,10))
point_size=150
point_border=0.8
plt.scatter(x[:,0],x[:,1],s=point_size,color="#00B3E9",edgecolors="black",lw=point_border)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
fig,ax=plt.subplots(figsize=figsize)
ep=0.5

for p in x:
    circle=plt.Circle(p,ep,color='#666666',fill=False,zorder=0,alpha=0.3)
    ax.add_artist(circle)
ax.scatter(x[:,0],x[:,1],s=point_size,c=pred,edgecolors="black",lw=point_border)
```
