```python
import plotly.express as px
plt.figure(figsize=(10,6))
fig = px.pie(insurance_data['sex'].value_counts().reset_index(), values='sex', names='index')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text="Gender Distribution of the Policy Holders")
fig.show()
```

```python
fig = px.pie(insurance_data['region'].value_counts().reset_index(), values='region', names='index')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text="Gender Distribution of the Policy Holders")
fig.show()
```

```python
fig = px.pie(insurance_data['smoker'].value_counts().reset_index(), values='smoker', names='index')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text="Distribution of Smokers or Non-Smokers")
fig.show()
```

```python
#Now lets create a boxplot to see the distribution of data for the charges when it comes to smokers and non-smokers 

plt.figure(figsize=(10,6))
fig = px.box(insurance_data, x='smoker', y='charges', color = 'sex')
fig.update_layout(title_text="Distribution of Charges Data")
fig.show()
```
