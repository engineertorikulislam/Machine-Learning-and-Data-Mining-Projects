```python
df = pd.read_csv("gender_classification.csv")
df.head()
```

```python
x_inputs = df.drop(['gender'],axis='columns')
x_inputs
```

```python
y_inputs = df[["gender"]]
y_inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y_inputs.gender = le.fit_transform(y_inputs.gender)
y_inputs
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_inputs,y_inputs, test_size=0.2)
```

```python
X_train
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
